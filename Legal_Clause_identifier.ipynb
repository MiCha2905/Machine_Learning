```python
def normalize_name(x):
    return str(x).lower().replace(".txt", "").replace(".pdf", "").strip()

# Make sure both dataframes have it
df["contract_norm"] = df["contract"].apply(normalize_name)
labels_df["contract_norm"] = labels_df["contract"].apply(normalize_name)

print("✅ Added 'contract_norm' column to both dataframes")
print("DF columns:", df.columns.tolist())
print("LABEL DF columns:", labels_df.columns.tolist())

print("Unique contracts (df):", len(df["contract_norm"].unique()))
print("Unique contracts (labels_df):", len(labels_df["contract_norm"].unique()))
```

```python
!pip install -q sentence-transformers
from sentence_transformers import SentenceTransformer, util
import torch, pandas as pd
from tqdm import tqdm

# Load a small but accurate sentence embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")  # use "nlpaueb/legal-bert-base-uncased" for legal domain
```

```python
matched_records = []

# Only match contracts that appear in both sets
common_contracts = set(df["contract_norm"]) & set(labels_df["contract_norm"])
print(f"✅ Found {len(common_contracts)} overlapping contracts.")

for contract in tqdm(common_contracts, desc="Semantic clause matching"):
    df_sub = df[df["contract_norm"] == contract]
    lbl_sub = labels_df[labels_df["contract_norm"] == contract]

    if lbl_sub.empty or df_sub.empty:
        continue

    # Encode both sides
    df_emb = model.encode(df_sub["clause_text"].tolist(), convert_to_tensor=True, show_progress_bar=False)
    lbl_emb = model.encode(lbl_sub["clause_text"].tolist(), convert_to_tensor=True, show_progress_bar=False)

    # Compute cosine similarities
    cosine_scores = util.cos_sim(df_emb, lbl_emb)

    for i, row in enumerate(df_sub.itertuples()):
        best_idx = torch.argmax(cosine_scores[i]).item()
        best_score = cosine_scores[i][best_idx].item()

        if best_score > 0.75:  # semantic similarity threshold
            matched_records.append({
                "contract": row.contract,
                "clause_text": row.clause_text,
                "clause_type_label": lbl_sub.iloc[best_idx]["clause_type_label"],
                "similarity": best_score
            })

matched_df = pd.DataFrame(matched_records)
print(f"✅ Matched clauses (semantic): {len(matched_df)}")
matched_df.head(5)
```

```python
# Complete upload -> extract -> classify -> structured output app (Colab-ready)
# 1 cell: install deps, load models, define functions, launch Gradio app.

# Install (run once)
!pip install -q sentence-transformers pdfplumber python-docx gradio rapidfuzz joblib spacy dateparser
!python -m spacy download en_core_web_sm

# ---- Imports ----
import os, io, re, json, zipfile
from pathlib import Path
from typing import List, Dict
import pandas as pd
import pdfplumber
from docx import Document
import dateparser
import joblib
from sentence_transformers import SentenceTransformer, util
import numpy as np
import spacy
from tqdm import tqdm

# Load spaCy NER (lighter & reliable)
nlp = spacy.load("en_core_web_sm")

# ---- Try loading models saved from your previous runs ----
clf_path = "/content/clause_classifier.pkl"
emb_path = "/content/sentence_transformer_model"

clf = None
emb_model = None

# load classifier
if os.path.exists(clf_path):
    try:
        clf = joblib.load(clf_path)
        print("Loaded classifier from", clf_path)
    except Exception as e:
        print("Could not load classifier from", clf_path, " — error:", e)

# load embedding model (try joblib saved; else load by name)
if os.path.exists(emb_path):
    try:
        emb_model = joblib.load(emb_path)
        print("Loaded embedding model from", emb_path)
    except Exception as e:
        print("Could not load embedding model from", emb_path, " — error:", e)

if emb_model is None:
    # fallback: load a sentence-transformer (fast & small)
    print("Loading default embedding model 'all-MiniLM-L6-v2' (this will download if not present)...")
    emb_model = SentenceTransformer("all-MiniLM-L6-v2")

if clf is None:
    raise FileNotFoundError(f"Classifier not found at {clf_path}. Put your trained classifier at this path (joblib .pkl).")

# If classifier supports predict_proba, we'll use it to compute confidence
has_proba = hasattr(clf, "predict_proba")

# ---- Text extraction utilities ----
def extract_text_from_pdf(file_obj) -> str:
    text_parts = []
    with pdfplumber.open(file_obj) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text() or ""
            text_parts.append(page_text + "\n")
    return "\n".join(text_parts)

def extract_text_from_docx(file_obj) -> str:
    doc = Document(file_obj)
    return "\n".join([p.text for p in doc.paragraphs])

def extract_text_from_uploaded(file) -> str:
    name = getattr(file, "name", "") or str(file)
    name = name.lower()
    # file may be a path or file-like object
    if name.endswith(".pdf"):
        return extract_text_from_pdf(file)
    elif name.endswith(".docx"):
        return extract_text_from_docx(file)
    else:
        # treat as text
        try:
            raw = file.read()
            if isinstance(raw, bytes):
                return raw.decode("utf-8", errors="ignore")
            return str(raw)
        except:
            # fallback: try opening as path
            with open(file, "r", encoding="utf-8", errors="ignore") as f:
                return f.read()

# ---- Clause splitter (robust for contracts) ----
def split_into_clauses(text: str) -> List[str]:
    # Normalize whitespace
    text = re.sub(r'\r\n', '\n', text)
    text = re.sub(r'\n{2,}', '\n\n', text)
    # split on double newlines, or numbered sections like "1. " "1.1 " or headings in ALL CAPS
    parts = re.split(r'(?:\n\s*\n|(?:(?<=\n)|^)\d+\.\s+)|(?=(?:[A-Z][A-Z\s]{10,}\n))', text)
    # fallback: split on sentences if nothing else
    clauses = [p.strip() for p in parts if len(p.strip())>30]
    # as a safety, if too many huge chunks, split on single newlines
    if len(clauses) < 10:
        clauses = [p.strip() for p in re.split(r'\n', text) if len(p.strip())>40]
    return clauses

# ---- Entity extraction (spaCy + regex) ----
money_re = re.compile(r'(\$|£|€)\s?\d+[,\d]*(?:\.\d{1,2})?')
date_word_re = re.compile(r'\b(?:January|February|March|April|May|June|July|August|September|October|November|December)\s+\d{1,2},?\s*\d{0,4}\b', re.IGNORECASE)

def extract_entities_from_clause(clause: str) -> Dict:
    doc = nlp(clause)
    parties = list({ent.text.strip() for ent in doc.ents if ent.label_ in ("ORG","PERSON")})
    dates = list({ent.text.strip() for ent in doc.ents if ent.label_ == "DATE"})
    dates += list(set(re.findall(date_word_re, clause)))
    amounts = list(set(re.findall(money_re, clause)))
    # normalize dates using dateparser where possible (best-effort)
    norm_dates = []
    for d in dates:
        try:
            parsed = dateparser.parse(d)
            if parsed:
                norm_dates.append(parsed.date().isoformat())
            else:
                norm_dates.append(d)
        except:
            norm_dates.append(d)
    return {"parties": parties, "dates": norm_dates, "amounts": amounts}

# ---- Clause classification ----
def classify_clauses(clause_texts: List[str]) -> List[Dict]:
    # embed in batches (memory safe)
    batch_size = 64
    out = []
    for i in range(0, len(clause_texts), batch_size):
        batch = clause_texts[i:i+batch_size]
        embeddings = emb_model.encode(batch, convert_to_numpy=True, show_progress_bar=False)
        preds = clf.predict(embeddings)
        if has_proba:
            probs = clf.predict_proba(embeddings)
            confs = probs.max(axis=1)
        else:
            # approximate confidence using distance to decision boundary not available -> set 0.75 fallback
            confs = [0.75]*len(batch)
        for txt, p, c in zip(batch, preds, confs):
            out.append({"clause_text": txt, "pred_label": str(p), "confidence": float(c)})
    return out

# ---- Full processing of an uploaded contract ----
def process_uploaded_contract(file_obj, contract_name=None, min_clause_len=30):
    text = extract_text_from_uploaded(file_obj)
    clauses = split_into_clauses(text)
    structured = []
    clause_texts = [c for c in clauses if len(c.strip())>=min_clause_len]
    if not clause_texts:
        clause_texts = clauses[:200]  # fallback
    classified = classify_clauses(clause_texts)
    for idx, item in enumerate(classified):
        clause = item["clause_text"]
        ents = extract_entities_from_clause(clause)
        structured.append({
            "contract": contract_name or getattr(file_obj, "name", "uploaded_contract"),
            "clause_id": idx,
            "clause_text": clause,
            "clause_type": item["pred_label"],
            "confidence": round(item["confidence"], 4),
            "parties": ents["parties"],
            "dates": ents["dates"],
            "amounts": ents["amounts"]
        })
    df_out = pd.DataFrame(structured)
    return df_out

# ---- Export helpers ----
def df_to_docx(df: pd.DataFrame, out_path: str):
    doc = Document()
    doc.add_heading("Structured Contract Clauses", level=1)
    for _, row in df.iterrows():
        doc.add_heading(f"{row['clause_type']} (confidence: {row['confidence']})", level=3)
        doc.add_paragraph(row['clause_text'])
        doc.add_paragraph(f"Parties: {row['parties']}")
        doc.add_paragraph(f"Dates: {row['dates']}")
        doc.add_paragraph(f"Amounts: {row['amounts']}")
        doc.add_paragraph("\n")
    doc.save(out_path)
    return out_path

# ---- Gradio UI ----
import gradio as gr

def analyze_and_download(file):
    # process
    fname = getattr(file, "name", "uploaded")
    df_result = process_uploaded_contract(file, contract_name=fname)
    # save outputs
    csv_path = f"/content/structured_{Path(fname).stem}.csv"
    json_path = f"/content/structured_{Path(fname).stem}.json"
    docx_path = f"/content/structured_{Path(fname).stem}.docx"
    df_result.to_csv(csv_path, index=False)
    df_result.to_json(json_path, orient="records", indent=2)
    df_to_docx(df_result, docx_path)
    # prepare zip
    zip_path = f"/content/structured_{Path(fname).stem}.zip"
    with zipfile.ZipFile(zip_path, "w") as z:
        z.write(csv_path, os.path.basename(csv_path))
        z.write(json_path, os.path.basename(json_path))
        z.write(docx_path, os.path.basename(docx_path))
    # return table preview and zip file for download
    preview = df_result.head(50)
    return preview, zip_path

demo = gr.Interface(
    fn=analyze_and_download,
    inputs=gr.File(file_count="single", label="Upload contract (.pdf / .docx / .txt)"),
    outputs=[gr.Dataframe(label="Structured preview (first 50 rows)"), gr.File(label="Download structured ZIP (CSV, JSON, DOCX)")],
    title="Legal Contract → Clause Extractor & Classifier",
    description="Uploads are processed page-by-page (safe for 200+ pages). Outputs: clause-level structured CSV/JSON/DOCX with predicted clause type and entities."
)

demo.launch(share=True)

```

```python
# =====================================================
# 💼 Legal Contract Clause Extractor & Classifier (Improved)
# Upload -> Extract -> Classify -> Clean -> Structured Download
# =====================================================

# Install dependencies (run once)
!pip install -q sentence-transformers pdfplumber python-docx gradio rapidfuzz joblib spacy dateparser matplotlib openpyxl
!python -m spacy download en_core_web_sm

# ---- Imports ----
import os, re, json, zipfile, io
from pathlib import Path
from typing import List, Dict
import pandas as pd
import pdfplumber
from docx import Document
import dateparser
import joblib
from sentence_transformers import SentenceTransformer
import numpy as np
import spacy
import matplotlib.pyplot as plt
from tqdm import tqdm

# ---- Load Models ----
nlp = spacy.load("en_core_web_sm")

clf_path = "/content/clause_classifier.pkl"
emb_path = "/content/sentence_transformer_model"

clf = None
emb_model = None

# Load classifier
if os.path.exists(clf_path):
    try:
        clf = joblib.load(clf_path)
        print("✅ Loaded classifier from", clf_path)
    except Exception as e:
        print("⚠️ Could not load classifier:", e)
else:
    raise FileNotFoundError("❌ Classifier not found at /content/legal_clause_classifier.pkl")

# Load embedding model
if os.path.exists(emb_path):
    try:
        emb_model = joblib.load(emb_path)
        print("✅ Loaded embedding model from", emb_path)
    except Exception as e:
        print("⚠️ Could not load embedding model:", e)

if emb_model is None:
    print("⬇️ Loading default embedding model (all-MiniLM-L6-v2)...")
    emb_model = SentenceTransformer("all-MiniLM-L6-v2")

has_proba = hasattr(clf, "predict_proba")

# ---- Text extraction ----
def extract_text_from_pdf(file_obj) -> str:
    text_parts = []
    with pdfplumber.open(file_obj) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text() or ""
            text_parts.append(page_text + "\n")
    return "\n".join(text_parts)

def extract_text_from_docx(file_obj) -> str:
    doc = Document(file_obj)
    return "\n".join([p.text for p in doc.paragraphs])

def extract_text_from_uploaded(file) -> str:
    name = getattr(file, "name", "") or str(file)
    name = name.lower()
    if name.endswith(".pdf"):
        return extract_text_from_pdf(file)
    elif name.endswith(".docx"):
        return extract_text_from_docx(file)
    else:
        try:
            raw = file.read()
            if isinstance(raw, bytes):
                return raw.decode("utf-8", errors="ignore")
            return str(raw)
        except:
            with open(file, "r", encoding="utf-8", errors="ignore") as f:
                return f.read()

# ---- Clause splitting ----
def split_into_clauses(text: str) -> List[str]:
    text = re.sub(r'\r\n', '\n', text)
    text = re.sub(r'\n{2,}', '\n\n', text)
    parts = re.split(r'(?:\n\s*\n|(?:(?<=\n)|^)\d+\.\s+)|(?=(?:[A-Z][A-Z\s]{10,}\n))', text)
    clauses = [p.strip() for p in parts if len(p.strip()) > 30]
    if len(clauses) < 10:
        clauses = [p.strip() for p in re.split(r'\n', text) if len(p.strip()) > 40]
    return clauses

# ---- Entity extraction ----
money_re = re.compile(r'(\$|£|€)\s?\d+[,\d]*(?:\.\d{1,2})?')
date_word_re = re.compile(r'\b(?:January|February|March|April|May|June|July|August|September|October|November|December)\s+\d{1,2},?\s*\d{0,4}\b', re.IGNORECASE)

def extract_entities_from_clause(clause: str) -> Dict:
    doc = nlp(clause)
    parties = list({ent.text.strip() for ent in doc.ents if ent.label_ in ("ORG", "PERSON")})
    dates = list({ent.text.strip() for ent in doc.ents if ent.label_ == "DATE"})
    dates += list(set(re.findall(date_word_re, clause)))
    amounts = list(set(re.findall(money_re, clause)))

    # Normalize dates using dateparser
    norm_dates = []
    for d in dates:
        try:
            parsed = dateparser.parse(d)
            if parsed:
                norm_dates.append(parsed.date().isoformat())
            else:
                norm_dates.append(d)
        except:
            norm_dates.append(d)
    return {"parties": parties, "dates": norm_dates, "amounts": amounts}

# ---- Clause classification ----
def classify_clauses(clause_texts: List[str]) -> List[Dict]:
    out = []
    batch_size = 64
    for i in range(0, len(clause_texts), batch_size):
        batch = clause_texts[i:i + batch_size]
        embeddings = emb_model.encode(batch, convert_to_numpy=True, show_progress_bar=False)
        preds = clf.predict(embeddings)
        if has_proba:
            probs = clf.predict_proba(embeddings)
            confs = probs.max(axis=1)
        else:
            confs = [0.75] * len(batch)
        for txt, p, c in zip(batch, preds, confs):
            out.append({"clause_text": txt, "pred_label": str(p), "confidence": float(c)})
    return out

# ---- Cleaning utility ----
def clean_entities(df: pd.DataFrame) -> pd.DataFrame:
    import ast
    for col in ["parties", "dates", "amounts"]:
        df[col] = df[col].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith("[") else x)
    # Filter duplicates, remove numeric noise, flatten for readability
    df["dates"] = df["dates"].apply(lambda d: [x for x in set(d) if len(x) >= 8 and not x.isdigit()])
    df["parties"] = df["parties"].apply(lambda p: [x for x in set(p) if len(x.split()) <= 4])
    df["amounts"] = df["amounts"].apply(lambda a: [x for x in set(a)])
    # Flatten lists for cleaner export
    for col in ["parties", "dates", "amounts"]:
        df[col] = df[col].apply(lambda x: "; ".join(x) if isinstance(x, list) else x)
    return df

# ---- Full processing ----
def process_uploaded_contract(file_obj, contract_name=None, min_clause_len=30):
    text = extract_text_from_uploaded(file_obj)
    clauses = split_into_clauses(text)
    clause_texts = [c for c in clauses if len(c.strip()) >= min_clause_len]
    classified = classify_clauses(clause_texts)
    structured = []
    for idx, item in enumerate(classified):
        clause = item["clause_text"]
        ents = extract_entities_from_clause(clause)
        structured.append({
            "contract": contract_name or getattr(file_obj, "name", "uploaded_contract"),
            "clause_id": idx,
            "clause_text": clause,
            "clause_type": item["pred_label"],
            "confidence": round(item["confidence"], 4),
            "parties": ents["parties"],
            "dates": ents["dates"],
            "amounts": ents["amounts"]
        })
    df_out = pd.DataFrame(structured)
    df_out = clean_entities(df_out)
    return df_out

# ---- Export helpers ----
def df_to_docx(df: pd.DataFrame, out_path: str):
    doc = Document()
    doc.add_heading("Structured Contract Clauses", level=1)
    for _, row in df.iterrows():
        doc.add_heading(f"{row['clause_type']} (confidence: {row['confidence']})", level=3)
        doc.add_paragraph(row['clause_text'])
        doc.add_paragraph(f"Parties: {row['parties']}")
        doc.add_paragraph(f"Dates: {row['dates']}")
        doc.add_paragraph(f"Amounts: {row['amounts']}")
        doc.add_paragraph("\n")
    doc.save(out_path)
    return out_path

# ---- Visualization ----
def plot_clause_summary(df: pd.DataFrame, save_path: str):
    counts = df["clause_type"].value_counts()
    plt.figure(figsize=(6, 4))
    counts.plot(kind="bar")
    plt.title("Clause Type Distribution")
    plt.xlabel("Clause Type")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()

# ---- Gradio App ----
import gradio as gr

def analyze_and_download(file):
    fname = getattr(file, "name", "uploaded")
    df_result = process_uploaded_contract(file, contract_name=fname)

    csv_path = f"/content/structured_{Path(fname).stem}.csv"
    json_path = f"/content/structured_{Path(fname).stem}.json"
    xlsx_path = f"/content/structured_{Path(fname).stem}.xlsx"
    docx_path = f"/content/structured_{Path(fname).stem}.docx"
    chart_path = f"/content/summary_{Path(fname).stem}.png"

    df_result.to_csv(csv_path, index=False)
    df_result.to_json(json_path, orient="records", indent=2)
    df_result.to_excel(xlsx_path, index=False)
    df_to_docx(df_result, docx_path)
    plot_clause_summary(df_result, chart_path)

    # Package everything
    zip_path = f"/content/structured_{Path(fname).stem}.zip"
    with zipfile.ZipFile(zip_path, "w") as z:
        for f in [csv_path, json_path, xlsx_path, docx_path, chart_path]:
            z.write(f, os.path.basename(f))

    preview = df_result.head(50)
    return preview, chart_path, zip_path

demo = gr.Interface(
    fn=analyze_and_download,
    inputs=gr.File(file_count="single", label="📄 Upload contract (.pdf / .docx / .txt)"),
    outputs=[
        gr.Dataframe(label="🧾 Structured Preview (first 50 rows)"),
        gr.Image(label="📊 Clause Type Summary"),
        gr.File(label="⬇️ Download Structured ZIP (CSV, XLSX, JSON, DOCX, Chart)")
    ],
    title="⚖️ Legal Contract → Clause Extractor & Classifier (Enhanced)",
    description="Upload a contract. Automatically extract clauses, classify them, detect entities (dates, parties, amounts), and download cleaned structured results."
)

demo.launch(share=False)
```
